**Data Conversion**

Our Churn, International Plan and Voice Mail plan values are based in booleans of true-false and yes-no, the column would be more effective if the values were converted to boolean.

```python
#creating a function to convert the numeric columns
def numeric_conversion(data, columns):
    #looping the conversion function
    for col in columns:
        #converting the yes and no to ones and zeros
        data[col] = data[col].replace('yes', 1)
        data[col] = data[col].replace('no', 0)
    return data
```

```python
#creating a list of columns to be converted
conversions = ['international plan', 'voice mail plan']

#execting the function to the list
df = numeric_conversion(df, conversions)
```

```python
#looking for features that are correlated above 0.5
#creating an empty set to store correlation features
highly_correlated_features = set()

#iterating through each column in the matrix
for corr1 in range(len(correlation_matrix.columns)):
    
    #using a nested loop to compare correlations
    for corr2 in range(corr1):
        #checking if the absolute correlation value is greater than 0.5
        if abs(correlation_matrix.iloc[corr1, corr2]) > 0.5:
            #identifying the column name with high correlations
            colname = correlation_matrix.columns[corr1]
            #adding the column name to the highly correlated features
            highly_correlated_features.add(colname)

#dropping the highly correlated features
df.drop(highly_correlated_features, axis=1, inplace=True)
```

I am removing the correlations greater than 0.5 since the colinearity will affect the model performance and might not give accurate results.

# Modelling

In this process I am going to start the modelling which is finding a model that can give fair predictions on the dataset. 

```python
#importing necesary modules for modelling
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV 
import imblearn
from imblearn.over_sampling import SMOTE
from sklearn.metrics import f1_score,recall_score,precision_score,confusion_matrix,roc_curve,roc_auc_score,classification_report 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
import xgboost as xgb
from xgboost import XGBClassifier
```

## Model Preparation

To achieve the modelling process I will start by preparing the data I have for modelling the data will be prepared to boost the model performance and avoid errors that might be brought by unprepared data.

#### Label Encoding

I will use the label encoding to convert categorical data into numerical data. The purpose of this is we need numerical data to drive through the machine learning models. In this case I will convert the True and False into binary of 0s and 1s.

```python
#initializing the label encoder
label_encoder = LabelEncoder()

#converting the columns with 'yes' or 'no' to 1 and 0
df['churn'] = label_encoder.fit_transform(df['churn'])
```

```python
#confirming the changes
df['churn'].value_counts()
```

**One Hot Encoding**

We need modified data that converts categorical data into a more suitable format for machine learning. Each categorical feature in the data frame is expended into multiple binary columns representing different categories.

```python
#using the get dummies function to perform one hot encoding
data = pd.get_dummies(df, drop_first = True)
```

```python
#confirming the changes
data.head()
```

**Scaling**

I am going to implement the `MinMaxScaler()` function to normalize numerical features within a uniform range, between 0 to 1. This ensures that the attributes have consistent scales thus aiding machine learning algorithms in better interpretation and convergence.

```python
#initializing the scaling function
scaler = MinMaxScaler()
#listing the columns containing numerical data
numerical_cols = ['account length', 'total day minutes', 'total day calls', 'total eve minutes',
                  'total eve calls', 'total night minutes', 'total night calls', 'total intl calls',
                  'customer service calls_1', 'customer service calls_2', 'customer service calls_3', 
                  'customer service calls_4', 'customer service calls_5', 'customer service calls_6+']
#fitting the data and transorming it into scaled versions
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

```python
data.head()
```

**Training the Model**

I am dividing the model into data that is trained and data that will test the model. The training 70% of the data is used to fit in the machine learning model. The testing data is used to evaluate the model's performance. This will enable a more accurate estimation of the model's predictive abilities.

```python
#creating the feature X and y variables for modelling
X = data.drop(['churn'], axis=1)
y = data['churn']

#separation of data between training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**SMOTE application**

I am going to apply SMOTE which is an oversampling technique in this case it will help overcome the prolem of overfitting caused by random oversampling. This will create new instances that help with interpolation between positive instances that are together.

```python
sm = SMOTE()

X_train_resampled, y_train_resampled = sm.fit_resample(X_train, y_train)
X_test_resampled, y_test_resampled = sm.fit_resample(X_test, y_test)
```

```python
print(pd.Series(y_train).value_counts())
print(pd.Series(y_train_resampled).value_counts())
print(pd.Series(y_test).value_counts())
print(pd.Series(y_test_resampled).value_counts())
```

# Modelling

## 1. Logistic Regression

The first model I'll use is logistic regression because churn is binary and logistic regression works well with binary classification. Logistic regression is specifically designed for binary classification tasks where the target variable has two possible outcomes (churn or not churn).

```python
#instantiate the logistic regression
logreg = LogisticRegression(random_state=123)
```

```python
#importing the warnings module
import warnings
#filtering the warnings
warnings.filterwarnings('ignore')

#fitting the model on the training data 
logreg.fit(X_train_resampled, y_train_resampled)
#predicting on the labels of test set
y_pred_log = logreg.predict(X_test)
```

There were warnings which were affecting the notebook's readability and neatness therefore I imported a feature called future warnings to help filter them out. 

A confusion matrix is a table used to evaluate the performance of a classification model. It helps to understand the model's predictions in terms of true positives, true negatives, false positives, and false negatives. The confusion matrix would be effective in our analysis of model performance.

```python
conf_matrix = confusion_matrix(y_test, y_pred_log)
print("Confusion Matrix:")
print(conf_matrix)
```

So we have:

**762** instances were correctly predicted as non-churn `True Negatives`.

**89** instances were correctly predicted as churn `True Positives`.

**95** instances were predicted as churn but actually were non-churn `False Positives`.

**54** instances were predicted as non-churn but actually were churn `False Negatives`.

I did a confusion matrix to help guage the data then realised that a heatmap would be better suited for this format. The heatmap assists on not only seeing the classification but also visualization.

```python
#calculating confusion matrix on the test
cm = confusion_matrix(y_test, y_pred_log)

#creating a heatmap visualization
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')

#showing the plot
plt.show()
```

So we have:

**Feature importance** allows us to pinpoint the most influential factors driving the model's predictions. It helps understand which attributes have more priority in determining the outcome (churn or non-churn). Understanding feature importance validates whether the model's predictions align with domain knowledge or intuition.

```python
#extracting feature importance coefficients from the LR model
importance = logreg.coef_[0]
#obtaining the names of the features used in the model
feature_names = X_train_resampled.columns
#creating a Series with feature importance values indexed by feature names
feature_importances = pd.Series(importance,index=feature_names)
#sorting feature importances in descending order to identify the most important features
feature_importances = feature_importances.sort_values(ascending=False)

#creating a visualization of the top 10 most important features
plt.figure(figsize=(10, 6))
top_features = feature_importances[:10]  
top_features.sort_values().plot(kind='barh')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.title('Top 10 Feature Importances')
plt.xlim(0, max(top_features)* 1.1) 
plt.show()
```

### ROC Curve

The ROC curve is a graphical representation used to evaluate the performance of a classification model across different thresholds. It plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold settings.

The ROC curve allows comparison of multiple models' performance on a single graph. Each model's curve (Logistic Regression, Random Forest, Decision Tree, and XGBoost) represents its ability to distinguish between classes across different threshold values.

```python
#setting a random seed
np.random.seed(123)
#listing classifiers to be evaluated
classifiers = [LogisticRegression(),
               RandomForestClassifier(),
               DecisionTreeClassifier(),
               XGBClassifier()]


#resulting table as a DataFrame
result_table = pd.DataFrame(columns=['classifiers', 'fpr','tpr','auc'])
```

```python
#trainijng the models and record the results
for cls in classifiers:
    #fitting the classifier on the resampled training data
    model = cls.fit(X_train_resampled, y_train_resampled)
    #predicting probabilities for the test set
    yproba = model.predict_proba(X_test)[::,1]
    
    #computing false positive rate, true positive rate, and AUC-ROC score
    fpr, tpr, _ = roc_curve(y_test,  yproba)
    auc = roc_auc_score(y_test, yproba)
    
    #appending classifier results to the result table
    result_table = result_table.append({'classifiers':cls.__class__.__name__,
                                        'fpr':fpr, 
                                        'tpr':tpr, 
                                        'auc':auc}, ignore_index=True)
```

```python
# Set name of the classifiers as index labels
result_table.set_index('classifiers', inplace=True)
#generating a figure
fig = plt.figure(figsize=(8,6))

#plotting the ROC curve for each classifier stored in the result table
for i in result_table.index:
    plt.plot(result_table.loc[i]['fpr'], 
             result_table.loc[i]['tpr'], 
             label="{}, AUC={:.3f}".format(i, result_table.loc[i]['auc']))

#plotting a line representing a random classifier
plt.plot([0,1], [0,1], color='orange', linestyle='--')
#generating the x ticks and label
plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("False Positive Rate", fontsize=15)
#generating the y ticks and label
plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("True Positive Rate", fontsize=15)
#generating the title and legend
plt.title('ROC Curve Analysis', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')
#showing the plot
plt.show()
```

Tuning our model is essential for a few common factors:

### XG boost

Considering that the best performing model was XG boost we will tune the model to gain the final results.

```python
#defining parameters for tuning
params = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.1, 0.01],
    'n_estimators': [100, 1000],   
    'subsample': [0.5, 0.8, 1.0]
}
```

```python
#creating an XGBoost classifier instance
xg_reg = xgb.XGBClassifier()
#setting up GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(xg_reg, param_grid=params, scoring='roc_auc',
                           cv=5, n_jobs=-1, verbose=1)  
#fitting the grid search to find the best parameters
grid_result = grid_search.fit(X_train_resampled, y_train_resampled)  
```

```python
#finding the best score and best parameters from the grid search results
best_score = grid_result.best_score_
best_params = grid_result.best_params_

# Displaying the best score and corresponding best parameters
print("Best AUC: {:.3f}".format(best_score))
print("Best Parameters:", best_params)
```

The **AUC** score is a measure of how well the model distinguishes between classes. Considering the score is close to 1 it means that the model is doing well.

The best parameters represent the hyperparameter combinations used to come up with the best results.

```python
#creating an XGBoost classifier with the best hyperparameters obtained from grid search
model = xgb.XGBClassifier(n_estimators=best_params['n_estimators'], 
                          max_depth=best_params['max_depth'],
                          learning_rate=best_params['learning_rate'], 
                          subsample=best_params['subsample'])

#fitting the model on the resampled dataset
model.fit(X_train_resampled, y_train_resampled)
```
